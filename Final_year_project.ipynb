```python
# Basic CNN


import os  #This library provides a portable way of using operating system dependent functionality. If you just want to read or write a file see open().
import numpy as np #NumPy is a Python library used for working with arrays. It also has functions for working in domain of linear algebra, fourier transform, and matrices.
import cv2 #commonly used functions and methods in CV2, including image loading and display, image manipulation, and image filtering.
import matplotlib.pyplot as plt #Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.
from sklearn.model_selection import train_test_split #To split the data into train and test directories.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.utils import to_categorical


# Set image size for resizing
IMG_SIZE = 128

# Function to load images from the dataset
def load_images(dataset_dir):
    images = []
    labels = []

    for category in ['genuine', 'forged']:
        label = 0 if category == 'genuine' else 1
        folder = os.path.join(dataset_dir, category)

        for img_name in os.listdir(folder):
            img_path = os.path.join(folder, img_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))  # Resize image
            img = img_to_array(img) / 255.0  # Normalize
            images.append(img)
            labels.append(label)

    return np.array(images), np.array(labels)

# Load the dataset
train_dir = 'signatures/train'
test_dir = 'signatures/test'

X_train, y_train = load_images(train_dir)
X_test, y_test = load_images(test_dir)

# One-hot encode the labels
y_train = to_categorical(y_train, 2)
y_test = to_categorical(y_test, 2)

# Define the CNN model
def create_model():
    model = Sequential()
    model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 1)))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Conv2D(128, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Flatten())
    model.add(Dense(128, activation='relu')) #
    model.add(Dropout(0.5))
    model.add(Dense(2, activation='softmax'))  # 2 classes: genuine and forged

    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    return model

# Create the model
model = create_model()

# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")

# Plot training and validation accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Define paths to original directories for genuine and forged signatures
original_genuine_dir = 'signatures/full_org'
original_forged_dir = 'signatures/full_forg'

# Define directories for the split dataset
split_dataset_dir = 'signatures'
train_genuine_dir = os.path.join(split_dataset_dir, 'train/genuine')
test_genuine_dir = os.path.join(split_dataset_dir, 'test/genuine')
train_forged_dir = os.path.join(split_dataset_dir, 'train/forged')
test_forged_dir = os.path.join(split_dataset_dir, 'test/forged')

# Create directories if they don't exist
os.makedirs(train_genuine_dir, exist_ok=True)
os.makedirs(test_genuine_dir, exist_ok=True)
os.makedirs(train_forged_dir, exist_ok=True)
os.makedirs(test_forged_dir, exist_ok=True)

# Function to split and move files
def split_and_move_files(file_list, train_dir, test_dir, train_files, test_files):
    for file in train_files:
        shutil.copy(file, train_dir)
    for file in test_files:
        shutil.copy(file, test_dir)

# Function to split data for genuine and forged signatures
def split_data(source_dir, train_dir, test_dir, test_size=0.2):
    # List all image files in the source directory
    all_files = [os.path.join(source_dir, f) for f in os.listdir(source_dir) if f.endswith('.png') or f.endswith('.jpg')]

    # Split the files into train and test sets
    train_files, test_files = train_test_split(all_files, test_size=test_size, random_state=42)

    # Move files to their respective directories
    split_and_move_files(all_files, train_dir, test_dir, train_files, test_files)

# Split genuine and forged signature images
split_data(original_genuine_dir, train_genuine_dir, test_genuine_dir)
split_data(original_forged_dir, train_forged_dir, test_forged_dir)

print("Data successfully split into train and test directories.")
```

```python
# Data augmentation


import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import img_to_array, ImageDataGenerator
from tensorflow.keras.utils import to_categorical

# Set image size for resizing
IMG_SIZE = 128

# Function to load images from the dataset
def load_images(dataset_dir):
    images = []
    labels = []

    for category in ['genuine', 'forged']:
        label = 0 if category == 'genuine' else 1
        folder = os.path.join(dataset_dir, category)

        for img_name in os.listdir(folder):
            img_path = os.path.join(folder, img_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))  # Resize image
            img = img_to_array(img) / 255.0  # Normalize
            images.append(img)
            labels.append(label)

    return np.array(images), np.array(labels)

# Load the dataset
train_dir = 'signatures/train'
test_dir = 'signatures/test'

X_train, y_train = load_images(train_dir)
X_test, y_test = load_images(test_dir)

# One-hot encode the labels
y_train = to_categorical(y_train, 2)
y_test = to_categorical(y_test, 2)

# Define the CNN model
def create_model():
    model = Sequential()
    model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 1)))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Conv2D(128, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Flatten())
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(2, activation='softmax'))  # 2 classes: genuine and forged

    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    return model

# Data augmentation
datagen = ImageDataGenerator(
    rotation_range=10,        # Rotate images up to 10 degrees
    width_shift_range=0.1,    # Shift images horizontally by 10%
    height_shift_range=0.1,   # Shift images vertically by 10%
    shear_range=0.1,          # Shear transformation
    zoom_range=0.1,           # Zoom in/out
    horizontal_flip=False,    # Keep as False for signature
    fill_mode='nearest'       # Filling strategy for augmented pixels
)

# Apply data augmentation to training data
datagen.fit(X_train)

# Create the model
model = create_model()

# Train the model using data augmentation
history = model.fit(
    datagen.flow(X_train, y_train, batch_size=32),
    epochs=10,
    validation_data=(X_test, y_test)
)

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")

# Plot training and validation accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Reshape
from tensorflow.keras.utils import to_categorical
from spektral.layers import GraphConv
from spektral.data import Graph

# Set default image size for resizing
IMG_SIZE = 128

# Function to load images from the dataset
def load_images(dataset_dir, img_size=IMG_SIZE):
    images = []
    labels = []
    for category in ['genuine', 'forged']:
        label = 0 if category == 'genuine' else 1
        folder = os.path.join(dataset_dir, category)
        
        if not os.path.exists(folder):
            print(f"Warning: The directory '{folder}' does not exist. Skipping this category.")
            continue

        for img_name in os.listdir(folder):
            img_path = os.path.join(folder, img_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            img = cv2.resize(img, (img_size, img_size))  # Resize image
            img = img / 255.0  # Normalize
            images.append(img)
            labels.append(label)

    images = np.array(images)
    images = np.expand_dims(images, -1)  # Add channel dimension for grayscale
    return images, np.array(labels)

# Load adjacency matrix for GNN layer
def build_adjacency_matrix(n_nodes):
    # Simple adjacency matrix for fully connected graph
    adj_matrix = np.ones((n_nodes, n_nodes)) - np.eye(n_nodes)
    return adj_matrix

# Set dataset paths
train_dir = input("signatures/train")
test_dir = input("signatures/test")

# Load the dataset
X_train, y_train = load_images(train_dir)
X_test, y_test = load_images(test_dir)

if X_train.size == 0 or y_train.size == 0:
    print("Error: Training data not loaded. Check the directory paths.")
    exit()
if X_test.size == 0 or y_test.size == 0:
    print("Error: Testing data not loaded. Check the directory paths.")
    exit()

# One-hot encode the labels
y_train = to_categorical(y_train, 2)
y_test = to_categorical(y_test, 2)

# Build adjacency matrix for GNN layer
n_nodes = 128  # Define the number of graph nodes
adj_matrix = build_adjacency_matrix(n_nodes)

# Define the model with CNN and GNN
def create_gnn_model():
    # CNN layers for feature extraction
    cnn_input = Input(shape=(IMG_SIZE, IMG_SIZE, 1))
    x = Conv2D(32, (3, 3), activation='relu')(cnn_input)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = Conv2D(64, (3, 3), activation='relu')(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = Conv2D(128, (3, 3), activation='relu')(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = Flatten()(x)
    
    # Reshape for GNN input (as nodes and features)
    x = Dense(n_nodes, activation='relu')(x)
    x = Dropout(0.5)(x)
    x = Reshape((n_nodes, 1))(x)  # Reshape to (n_nodes, 1 feature per node)

    # GNN layer
    gnn_layer = GraphConv(64, activation='relu')([x, adj_matrix])
    
    # Fully connected layers after GNN
    x = Flatten()(gnn_layer)
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.5)(x)
    output = Dense(2, activation='softmax')(x)

    # Define model
    model = Model(inputs=cnn_input, outputs=output)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    
    return model

# Initialize the model
model = create_gnn_model()

# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy with GNN: {accuracy * 100:.2f}%")

# Plot training and validation accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs (CNN + GNN)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# ResNet50

import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam

# Directory paths for your dataset
train_dir = 'signatures/train'  # Training dataset path
valid_dir = 'signatures/test'  # Validation dataset path

# Image size for ResNet50
img_height, img_width = 224, 224  # ResNet50 input size
batch_size = 32

# Data generators for training and validation
train_datagen = ImageDataGenerator(rescale=1./255, 
                                   shear_range=0.2, 
                                   zoom_range=0.2, 
                                   horizontal_flip=True)
valid_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary'  # Binary classification (genuine or forged)
)

valid_generator = valid_datagen.flow_from_directory(
    valid_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary'  # Binary classification
)

# Load the ResNet50 model pre-trained on ImageNet, excluding the top layer
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))

# Freeze the base model to prevent it from being trained
base_model.trainable = False

# Build the model
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')  # Output layer for binary classification
])

# Compile the model
model.compile(optimizer=Adam(learning_rate=1e-4),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(
    train_generator,
    epochs=10,
    validation_data=valid_generator
)

# Save the model
model.save('signature_verification_model.h5')

# Evaluate the model on the validation data
validation_loss, validation_accuracy = model.evaluate(valid_generator)
print(f'Validation Loss: {validation_loss}')
print(f'Validation Accuracy: {validation_accuracy}')
```

```python
# CNN with GAN

import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten, LeakyReLU, Dropout, Reshape
from keras.optimizers import Adam
from keras.datasets import mnist
from keras import initializers

# Set random seed for reproducibility
np.random.seed(1000)

# Define constants
IMG_SIZE = 28
LATENT_DIM = 100
BATCH_SIZE = 64
EPOCHS = 10000
SAVE_INTERVAL = 1000

# Load MNIST dataset
(X_train, _), (_, _) = mnist.load_data()
X_train = X_train / 127.5 - 1.0  # Rescale to [-1, 1]
X_train = X_train.reshape(X_train.shape[0], IMG_SIZE, IMG_SIZE, 1)

# Create the discriminator model
def create_discriminator():
    model = Sequential()
    model.add(Conv2D(64, (3, 3), strides=(2, 2), padding='same', input_shape=(IMG_SIZE, IMG_SIZE, 1)))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Dropout(0.4))
    
    model.add(Conv2D(128, (3, 3), strides=(2, 2), padding='same'))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Dropout(0.4))
    
    model.add(Flatten())
    model.add(Dense(1, activation='sigmoid'))  # Output: real (1) or fake (0)
    
    model.compile(loss='binary_crossentropy', optimizer=Adam(), metrics=['accuracy'])
    return model

# Create the generator model
def create_generator():
    model = Sequential()
    
    model.add(Dense(128 * 7 * 7, activation='relu', input_dim=LATENT_DIM))
    model.add(Reshape((7, 7, 128)))
    model.add(Conv2D(128, (3, 3), padding='same'))
    model.add(LeakyReLU(alpha=0.2))
    
    model.add(Conv2D(64, (3, 3), padding='same'))
    model.add(LeakyReLU(alpha=0.2))
    
    model.add(Conv2D(1, (3, 3), padding='same', activation='tanh'))
    return model

# Create the GAN model
def create_gan(generator, discriminator):
    discriminator.trainable = False
    model = Sequential()
    model.add(generator)
    model.add(discriminator)
    model.compile(loss='binary_crossentropy', optimizer=Adam())
    return model

# Training the GAN
def train_gan(epochs, batch_size, save_interval):
    # Labels for real and fake images
    valid = np.ones((batch_size, 1))  # Real images label
    fake = np.zeros((batch_size, 1))  # Fake images label
    
    # Create models
    discriminator = create_discriminator()
    generator = create_generator()
    gan = create_gan(generator, discriminator)
    
    # Training loop
    for epoch in range(epochs):
        # Train discriminator
        # Select a random batch of real images
        idx = np.random.randint(0, X_train.shape[0], batch_size)
        real_images = X_train[idx]
        
        # Generate a batch of fake images
        noise = np.random.normal(0, 1, (batch_size, LATENT_DIM))
        fake_images = generator.predict(noise)
        
        # Train the discriminator on real and fake images
        d_loss_real = discriminator.train_on_batch(real_images, valid)
        d_loss_fake = discriminator.train_on_batch(fake_images, fake)
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)
        
        # Train the generator (via the GAN model)
        noise = np.random.normal(0, 1, (batch_size, LATENT_DIM))  # Generate new noise
        g_loss = gan.train_on_batch(noise, valid)  # We want the generator to fool the discriminator
        
        # Print progress
        if epoch % save_interval == 0:
            print(f"{epoch} [D loss: {d_loss[0]} | D accuracy: {100 * d_loss[1]}] [G loss: {g_loss}]")
        
        # Save generated images at intervals (optional)
        if epoch % save_interval == 0:
            save_generated_images(epoch, generator)

# Function to save generated images
def save_generated_images(epoch, generator, examples=10, dim=(1, 10), figsize=(10, 1)):
    noise = np.random.normal(0, 1, (examples, LATENT_DIM))
    generated_images = generator.predict(noise)
    generated_images = (generated_images + 1) / 2.0  # Rescale to [0, 1]
    
    plt.figure(figsize=figsize)
    for i in range(examples):
        plt.subplot(dim[0], dim[1], i + 1)
        plt.imshow(generated_images[i, :, :, 0], cmap='gray')
        plt.axis('off')
    plt.tight_layout()
    plt.savefig(f"gan_generated_image_epoch_{epoch}.png")
    plt.close()

# Train the GAN
train_gan(epochs=10000, batch_size=BATCH_SIZE, save_interval=1000)
```
